```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['type'] = le.fit_transform(df['type'])
```

```python
X= df[['hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic','catsize','animal name']]
Y= df['type']
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_SS = SS.fit_transform(X)
X_SS = pd.DataFrame(X_SS)
X_SS.columns = list(X)
X_SS.head()
```

```python
# data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(X_train.values,Y_train)
Y_pred_train = model.predict(X_train.values)
Y_pred_test = model.predict(X_test.values)
                    
```

```python
# model metrics
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score, confusion_matrix
ac1 = accuracy_score(Y_train,Y_pred_train)
print("Training Accuracy score:", ac1.round(2))
ac2 = accuracy_score(Y_test,Y_pred_test)
print("Test Accuracy score:", ac2.round(2))
```

```python
cm = confusion_matrix(Y_train,Y_pred_train)
cm
```

```python
rs = recall_score(Y_train,Y_pred_train,average='micro')
print("sensitivity score:",rs.round(3))
```

```python
TN =cm[0,0]
FP = cm[0,1]
TNR = TN/(TN+FP)
print("specificity score:", TNR.round(2))


ps = precision_score(Y_train,Y_pred_train,average='micro')
print("precision score:",ps.round(3))
```

```python
f1score =f1_score(Y_train,Y_pred_train,average='micro')
print("f1 score:",f1score.round(3))
```

```python
# validation set method
```

```python
training_acc = []
test_acc = []
```

```python
for i in range(1,101,1):
    X_train,X_test,Y_train,Y_test = train_test_split(X_SS,Y,test_size=0.30,random_state=i)
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(X_train.values,Y_train) 
    Y_pred_train = model.predict(X_train.values)
    Y_pred_test = model.predict(X_test.values)
    training_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))
```

```python
print("Cross validation-training accuracy",np.mean(training_acc).round(2))
print("Cross validation-test accuracy",np.mean(test_acc).round(2))
```

```python
training_accuracy_list = []
test_accuracy_list = []
```

```python
for k in range(3,18 , 2):
    training_accuracy = []
    test_accuracy = []

    for i in range(1, 100, 1):
        X_train, X_test, Y_train, Y_test = train_test_split(X_SS, Y, test_size=0.20, random_state=i)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train.values, Y_train)
        y_pred_train = knn.predict(X_train.values)
        y_pred_test = knn.predict(X_test.values)
        training_accuracy.append(accuracy_score(Y_train, y_pred_train))
        test_accuracy.append(accuracy_score(Y_test, y_pred_test))
```

```python
training_accuracy_list.append(np.mean(training_accuracy).round(2))
test_accuracy_list.append(np.mean(test_accuracy).round(2))
```

```python
print("Training Accuracies:", training_accuracy_list)
print("Test Accuracies:", test_accuracy_list)
```

```python
import matplotlib.pyplot as plt

x_values = range(3, 18, 2)
plt.scatter(x_values[:len(training_accuracy_list)], training_accuracy_list, color='blue')
plt.plot(x_values[:len(training_accuracy_list)], training_accuracy_list, color='black')

plt.scatter(x_values[:len(test_accuracy_list)], test_accuracy_list, color='red')
plt.plot(x_values[:len(test_accuracy_list)], test_accuracy_list, color='black')

plt.show()
```

# 1. What are the key hyperparameters in KNN?

```python
# A. The key hyperparameters in KNN are:
```

# k:

```python
#The value of k determines how many neighbors to consider when making a prediction.
#   A larger value of k will result in a smoother decision boundary, but it may also lead to overfitting.
#   A smaller value of k will result in a more jagged decision boundary, but it may also lead to underfitting.
```

# distance_metric:

```python
#The distance metric determines how the distance between data points is calculated.
#   The most common distance metric is Euclidean distance, but other metrics such as Manhattan distance and cosine similarity can also be used.
```

# weights:

```python
#The weights determine how much each neighbor contributes to the prediction.
#   Uniform weights give each neighbor the same weight, while distance-based weights give closer neighbors more weight.
```

# 2. What distance metrics can be used in KNN?

```python
#The most common distance metric used in KNN is Euclidean distance.
# Euclidean distance is calculated as the square root of the sum of the squared differences between the values of each dimension.
# For example, the Euclidean distance between two points in two dimensions is calculated as follows:

# distance = sqrt((x1 - x2)^2 + (y1 - y2)^2)




# The following distance metrics can be used in KNN:
# * Manhattan distance
# * Cosine similarity
# * Minkowski distance

```
