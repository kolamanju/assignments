```python
X_cont = data1[['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']]
X_cate = data1[['Name','Sex','Ticket','Embarked']]
```

```python
# standardization 
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
X_cont_SS = SS.fit_transform(X_cont)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X_cont)
X_cont_SS.head()
```

# label encoding

```python

from sklearn.preprocessing import LabelEncoder
LE =LabelEncoder()
X_cont['Survived'] = LE.fit_transform(X_cont['Survived'])
X_cont.head()
```

```python
X= X_cont_SS
X
```

```python
Y =X_cont["Survived"]
Y
```

# data partition

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.5,random_state =20)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

# model metrics 

```python
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(Y_train,Y_pred_train)
print("training accuracy score:",ac1.round(3))
ac2 = accuracy_score(Y_test,Y_pred_test)
print("testing accuracy score:",ac2.round(3))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,r2_score,recall_score,precision_score,f1_score
cm = confusion_matrix(Y_train,Y_pred_train)
cm
```

```python
ac = accuracy_score(Y_train,Y_pred_train)
print("accuracy score:",ac.round(3))
```

```python
rs = recall_score(Y_train,Y_pred_train,average='micro')
print("sensitivity score:",rs.round(3))
```

```python
TN =cm[0,0]
FP = cm[0,1]
TNR = TN/(TN+FP)
print("specificity score:", TNR.round(2))


ps = precision_score(Y_train,Y_pred_train,average='micro')
print("precision score:",ps.round(3))
```

```python
f1score =f1_score(Y_train,Y_pred_train,average='micro')
print("f1 score:",f1score.round(3))
```

```python
pos_label = 1 
fpr, tpr,dummy= roc_curve(Y,  data1['proba'],pos_label=pos_label)
```

```python
proba = model.predict_proba(X) 

# Calculate AUC for each class
auc_scores = []
for class_idx in range(proba.shape[1]):
    auc = roc_auc_score(Y == class_idx, proba[:, class_idx])
    auc_scores.append(auc)
    print(f"AUC for class {class_idx}: {auc}")

# Calculate average AUC 
average_auc = np.mean(auc_scores)
print(f"Average AUC: {average_auc}")
```

#  Interpretation:

```python
model_coefficients = pd.DataFrame(model.coef_, columns=X_train.columns)
model_coefficients.index.name = 'Coefficient'
print(model_coefficients.T)
```

```python
for column, coefficient in model_coefficients.items():
  print(f"Feature: {column}, Coefficient: {coefficient[0]:.3f}")


sorted_features = model_coefficients.abs().sort_values(by=model_coefficients.columns[0], ascending=False)

print("\nTop 5 most significant features:")
print(sorted_features.index[:5].to_list())
```
