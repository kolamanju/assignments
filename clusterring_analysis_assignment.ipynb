```python
# K-Means clustering
# Elbow curve for different K values

sum_of_squared_distances = []
K = range(1, 15)
for k in K:
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(df)
  sum_of_squared_distances.append(kmeans.inertia_)

plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances')
plt.title('Elbow curve for K-Means clustering')
plt.show()
```

```python
# Hierarchical clustering
# Experiment with different linkage criteria

linkage_criteria = ['ward', 'complete', 'average', 'single']
for linkage in linkage_criteria:
  hc = AgglomerativeClustering(linkage=linkage, n_clusters=3)
  hc.fit(df)
  labels_hc = hc.labels_
  plt.scatter(df['Balance'], df['Qual_miles'], c=labels_hc, cmap='viridis')
  plt.title('Hierarchical Clustering with linkage: ' + linkage)
  plt.show()
```

```python
# DBSCAN clustering
# Experiment with different epsilon and minPts

eps_values = [0.5, 1.0, 1.5, 2.0]
min_pts_values = [5, 10, 15, 20]
for eps in eps_values:
  for min_pts in min_pts_values:
    dbscan = DBSCAN(eps=eps, min_samples=min_pts)
    dbscan.fit(df)
    labels_dbscan = dbscan.labels_
    plt.scatter(df['Balance'], df['Qual_miles'], c=labels_dbscan, cmap='viridis')
    plt.title('DBSCAN clustering with eps: ' + str(eps) + ' and minPts: ' + str(min_pts))
    plt.show()
```

```python
# K-Means clustering

# Analyze the clusters
kmeans_clusters = {}
for i in range(kmeans.n_clusters):
  cluster_members = df[labels_kmeans == i]
  kmeans_clusters[i] = {
      'size': len(cluster_members),
      'average_balance': cluster_members['Balance'].mean(),
      'average_qual_miles': cluster_members['Qual_miles'].mean()
  }

# Print insights
print("K-Means clustering insights:")
for i in kmeans_clusters:
  print(f"Cluster {i}:")
  print(f"\tSize: {kmeans_clusters[i]['size']}")
  print(f"\tAverage balance: {kmeans_clusters[i]['average_balance']}")
  print(f"\tAverage qual miles: {kmeans_clusters[i]['average_qual_miles']}")

```

```python
# DBSCAN clustering

# Analyze the clusters
dbscan_clusters = {}
for i in range(len(set(labels_dbscan))):
  cluster_members = df[labels_dbscan == i]
  dbscan_clusters[i] = {
      'size': len(cluster_members),
      'average_balance': cluster_members['Balance'].mean(),
      'average_qual_miles': cluster_members['Qual_miles'].mean()
  }

# Print insights
print("\nDBSCAN clustering insights:")
for i in dbscan_clusters:
  print(f"Cluster {i}:")
  print(f"\tSize: {dbscan_clusters[i]['size']}")
  print(f"\tAverage balance: {dbscan_clusters[i]['average_balance']}")
  print(f"\tAverage qual miles: {dbscan_clusters[i]['average_qual_miles']}")
```

```python
from sklearn.metrics import silhouette_score

# K-Means clustering
silhouette_score_kmeans = silhouette_score(df, labels_kmeans)
print(f"Silhouette score for K-Means clustering: {silhouette_score_kmeans}")
```

```python
# DBSCAN clustering
# Experiment with different epsilon and minPts

eps_values = [0.5, 1.0, 1.5, 2.0]
min_pts_values = [5, 10, 15, 20]
for eps in eps_values:
  for min_pts in min_pts_values:
    dbscan = DBSCAN(eps=eps, min_samples=min_pts)
    dbscan.fit(df)
    labels_dbscan = dbscan.labels_

    # Check if there are at least two clusters (excluding noise)
    n_clusters = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
    if n_clusters >= 2:
        plt.scatter(df['Balance'], df['Qual_miles'], c=labels_dbscan, cmap='viridis')
        plt.title('DBSCAN clustering with eps: ' + str(eps) + ' and minPts: ' + str(min_pts))
        plt.show()

        # Calculate silhouette score only if there are at least two clusters
        silhouette_score_dbscan = silhouette_score(df, labels_dbscan)
        print(f"Silhouette score for DBSCAN clustering (eps={eps}, minPts={min_pts}): {silhouette_score_dbscan}")
    else:
        print(f"DBSCAN with eps={eps} and minPts={min_pts} resulted in less than two clusters. Skipping silhouette score calculation.")
```
