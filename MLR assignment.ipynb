# VIF calculation 

```python
rsq_Age_08_04 = smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Cylinders+Weight', data=df1).fit().rsquared
vif_Age_08_04= 1/(1-rsq_Age_08_04)

rsq_KM = smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Cylinders+Weight', data=df1).fit().rsquared
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols('HP~Age_08_04+KM+cc+Doors+Gears+Cylinders+Weight', data=df1).fit().rsquared
vif_HP = 1/(1-rsq_HP)

rsq_cc = smf.ols('cc~Age_08_04+KM+HP+Doors+Gears+Cylinders+Weight', data=df1).fit().rsquared
vif_cc = 1/(1-rsq_cc)

rsq_Doors = smf.ols('Doors~Age_08_04+KM+HP+cc+Gears+Cylinders+Weight', data=df1).fit().rsquared
vif_Doors = 1/(1-rsq_Doors)

rsq_Gears = smf.ols('Gears~Age_08_04+KM+HP+cc+Doors+Cylinders+Weight', data=df1).fit().rsquared
vif_Gears = 1/(1-rsq_Gears)

rsq_Cylinders = smf.ols('Cylinders~Age_08_04+KM+HP+cc+Doors+Gears+Weight', data=df1).fit().rsquared
vif_Cylinders = 1/(1-rsq_Cylinders)

rsq_Weight = smf.ols('Weight~Age_08_04+KM+HP+cc+Doors+Gears+Cylinders', data=df1).fit().rsquared
vif_Weight = 1/(1-rsq_Weight)

d1 = {'variables' :['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears','Cylinders','Weight'], 'VIF': [vif_Age_08_04, vif_KM, vif_HP, vif_cc, vif_Doors, vif_Gears,vif_Cylinders, vif_Weight]}
vif_frame = pd.DataFrame(d1)
vif_frame
```

# Cook's Distance

```python
model_influence = model.get_influence()
(c,_)=model_influence.cooks_distance
fig = plt.subplots(figsize=(15,8))
plt.stem(np.arange(len(df1)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel;('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
