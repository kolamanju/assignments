1. Data Exploration and Preprocessing

```python
import pandas as pd
df=pd.read_csv("blogs.csv")
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
num_cols=df.shape[1]
```

```python

for i in range(1,num_cols):
    df.iloc[:,i] = LE.fit_transform(df.iloc[:,i])
```

```python
df.head()
df.info()
```

```python
Y = df.iloc[:,0]
X = df.iloc[:,1:]
```

```python
print(df.columns)
```

```python
Y = df.iloc[:, 0]
X = df.iloc[:, 1:]
Y.info()
Y.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X.iloc[:,0].astype(str))
print(X_tfidf.shape)
```

2. Naive Bayes Model for Text Classification

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_tfidf, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
Y_pred_train = model.predict(x_train)
Y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(y_train,Y_pred_train)
print("Training Accuracy score:", ac1.round(2))
ac2 = accuracy_score(y_test,Y_pred_test)
print("Test Accuracy score:", ac2.round(2))
```

3. Sentiment Analysis

```python
training_acc = []
test_acc = []
```

```python
from textblob import TextBlob

def get_sentiment(text):
  """
  Analyzes the sentiment of a given text and returns a string representing the sentiment ("positive", "negative", or "neutral").
  """
  analysis = TextBlob(text)
  polarity = analysis.sentiment.polarity
  if polarity > 0:
    return "positive"
  elif polarity < 0:
    return "negative"
  else:
    return "neutral"

df["Sentiment"] = df["Data"].apply(get_sentiment)
df.head()
```

```python
categories = ["News"] * 500 + ["Sports"] * 500 + ["Entertainment"] * 500 + ["Other"] * 500
df["Category"] = categories

sentiment_counts = df.groupby(["Category", "Sentiment"]).size().unstack()
print(sentiment_counts)
```

4. Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_train_pred = model.predict(x_train)
accuracy_train = accuracy_score(y_train, y_train_pred)
precision_train = precision_score(y_train, y_train_pred, average="weighted")
recall_train = recall_score(y_train, y_train_pred, average="weighted")
f1_train = f1_score(y_train, y_train_pred, average="weighted")
```

```python

y_test_pred = model.predict(x_test)
accuracy_test = accuracy_score(y_test, y_test_pred)
precision_test = precision_score(y_test, y_test_pred, average="weighted")
recall_test = recall_score(y_test, y_test_pred, average="weighted")
f1_test = f1_score(y_test, y_test_pred, average="weighted")
```

```python
print("Training set performance:")
print("- Accuracy:", accuracy_train)
print("- Precision:", precision_train)
print("- Recall:", recall_train)
print("- F1-score:", f1_train)
```
