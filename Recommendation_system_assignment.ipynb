INTERVIEW QUESTIONS

User-Based Collaborative Filtering

Concept:
User-based collaborative filtering makes recommendations based on the similarity between users. It assumes that users with similar past behaviors will have similar future preferences.


Steps:

Identify similar users: For a target user, identify other users who have similar preferences (based on ratings, interactions, etc.).

Generate recommendations: Recommend items that similar users have liked or interacted with but the target user has not yet experienced.


---

Item-Based Collaborative Filtering

Concept:
Item-based collaborative filtering makes recommendations based on the similarity between items. It assumes that items which received similar ratings from users will have similar characteristics.


Steps:


Identify similar items: For each item, find other items that have been rated similarly by users.

Generate recommendations: Recommend items that are similar to items the user has already liked or interacted with.



How Collaborative Filtering Works
Collaborative filtering works by collecting user interactions with items (like ratings, clicks, purchases, etc.) and using this data to identify patterns and relationships. There are two main types of collaborative filtering: user-based and item-based.

1. User-Based Collaborative Filtering

User-Item Interaction Matrix: Create a matrix where rows represent users and columns represent items. The entries in the matrix indicate the user's rating or interaction with an item.


Similarity Calculation: Calculate the similarity between users using measures like cosine similarity, Pearson correlation, etc.

Find Neighbors: Identify the most similar users (neighbors) to the target user.

Aggregate Preferences: Aggregate the preferences of the neighbors to make recommendations for the target user. Items that the neighbors liked but the target user hasn't interacted with are recommended.



2. Item-Based Collaborative Filtering

User-Item Interaction Matrix: Similar to user-based, create a matrix of user interactions.


Item Similarity Calculation: Calculate the similarity between items using measures like cosine similarity, Pearson correlation, etc.


Find Similar Items: Identify items that are similar to those the user has interacted with.


Aggregate Preferences: Recommend items similar to those the user has liked or interacted with in the past.
